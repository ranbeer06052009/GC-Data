```python
def convert_duration_to_seconds(duration_str):
    if pd.isna(duration_str) or duration_str == '0':
        return 0

    parts = str(duration_str).split(':')

    if len(parts) == 3:
        h, m, s = map(int, parts)
        return h * 3600 + m * 60 + s

    elif len(parts) == 2:
        m, s = map(int, parts)
        return m * 60 + s

    else:
        return float(duration_str)

# Convert duration columns
for col in ["Total Uploaded Duration", "Total Created Duration", "Total Published Duration"]:
    duration[col] = duration[col].apply(convert_duration_to_seconds)

# Interactive bar chart
fig = px.bar(
    duration,
    x="Month",
    y=[
        "Total Uploaded Duration",
        "Total Created Duration",
        "Total Published Duration"
    ],
    title="Monthly Processing Duration",
    barmode="group"
)

fig.update_layout(
    yaxis_title="Duration (seconds)",
    xaxis_title="Month",
    template="plotly_white"
)
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
duration_melt = duration.melt(
    id_vars="Month",
    value_vars=[
        "Total Uploaded Duration",
        "Total Created Duration",
        "Total Published Duration"
    ],
    var_name="Stage",
    value_name="Duration"
)

fig = px.bar(
    duration_melt,
    x="Month",
    y="Duration",
    color="Stage",
    title="Monthly Processing Duration"
)
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
uploaded = monthly["Total Uploaded"].sum()
processed = monthly["Total Created"].sum()
published = monthly["Total Published"].sum()

funnel = pd.DataFrame({
    "Stage": ["Total Uploaded", "Total Processed", "Total Published"],
    "Count": [uploaded, processed, published]
})

fig = px.bar(
    funnel,
    x="Stage",
    y="Count",
    title="Processing Funnel",
    text="Count"
)

fig.update_layout(
    template="plotly_white",
    yaxis_title="Number of Videos",
    xaxis_title="Processing Stage"
)

fig.show()

print("Publish Rate:", published / processed)
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Funnel(
    y = ["Total Uploaded","Total Processed","Total Published"],
    x = [uploaded, processed, published],
    textinfo = "value+percent initial"
))

fig.update_layout(title="Video Processing Funnel")

fig.show()
```

```python
publish_rate = (published / processed) * 100
print(f"Publish Rate: {publish_rate:.2f}%")
upload_to_process = processed / uploaded
process_to_publish = published / processed

print(f"Upload → Process Rate: {upload_to_process:.2%}")
print(f"Process → Publish Rate: {process_to_publish:.2%}")
```
