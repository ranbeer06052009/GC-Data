```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
monthly = pd.read_csv("/content/GC-Data/data/monthly-chart.csv")
duration = pd.read_csv("/content/GC-Data/data/month-wise-duration.csv")

channel_data = pd.read_csv("/content/GC-Data/data/CLIENT 1 combined_data(2025-3-1-2026-2-28).csv")

channel_user = pd.read_csv("/content/GC-Data/data/combined_data(2025-3-1-2026-2-28) by channel and user.csv")

input_type = pd.read_csv("/content/GC-Data/data/combined_data(2025-3-1-2026-2-28) by input type.csv")

language = pd.read_csv("/content/GC-Data/data/combined_data(2025-3-1-2026-2-28) by language.csv")

output_type = pd.read_csv("/content/GC-Data/data/combined_data(2025-3-1-2026-2-28) by output type.csv")

user_data = pd.read_csv("/content/GC-Data/data/combined_data(2025-3-1-2026-2-28) by user.csv")

platform = pd.read_csv("/content/GC-Data/data/channel-wise-publishing.csv")

platform_duration = pd.read_csv("/content/GC-Data/data/channel-wise-publishing duration.csv")

video_data = pd.read_csv("/content/GC-Data/data/video_list_data_obfuscated.csv")
```

```python
datasets = {
    "monthly": monthly,
    "duration": duration,
    "channel_data": channel_data,
    "channel_user": channel_user,
    "input_type": input_type,
    "language": language,
    "output_type": output_type,
    "user_data": user_data,
    "platform": platform,
    "platform_duration": platform_duration,
    "video_data": video_data
}

for name, df in datasets.items():
    print(f"\n{name.upper()} SHAPE:", df.shape)
    display(df.head())
```

```python
for df in datasets.values():
    df.columns = df.columns.str.strip()
```

```python
missing_summary = pd.DataFrame()

for name, df in datasets.items():
    missing = df.isnull().sum()
    missing_percent = (df.isnull().sum() / len(df)) * 100

    temp = pd.DataFrame({
        "dataset": name,
        "column": missing.index,
        "missing_count": missing.values,
        "missing_percent": missing_percent.values
    })

    missing_summary = pd.concat([missing_summary, temp])

missing_summary.sort_values("missing_percent", ascending=False).head(20)
```

```python
for name, df in datasets.items():
    dup = df.duplicated().sum()
    print(f"{name} duplicate rows:", dup)
```

```python
monthly.plot(x="Month", y=["Total Uploaded","Total Created","Total Published"], marker="o")
plt.title("Monthly Video Workflow Trend")
plt.ylabel("Number of Videos")
plt.show()
```

```python
duration.plot(x="Month",
              y=["Total Uploaded Duration","Total Created Duration","Total Published Duration"],
              kind="bar")

plt.title("Monthly Processing Duration")
plt.ylabel("Duration")
plt.show()
```

```python
uploaded = monthly["Total Uploaded"].sum()
processed = monthly["Total Created"].sum()
published = monthly["Total Published"].sum()

funnel = pd.DataFrame({
    "Stage":["Total Uploaded","Total Processed","Total Published"],
    "Count":[uploaded,processed,published]
})

sns.barplot(data=funnel, x="Stage", y="Count")
plt.title("Processing Funnel")
plt.show()

print("Publish Rate:", published/processed)
```
